The Confusion Matrix is the toevaluate the behavior and understanding the effectiveness of a binary or categorical classifier in machine learning.



```python
To calculate and visualize a confusion matrix, we took a simple example of DataFrame with two columns 'y_Actual', 'y_Predicted'.
```

```python
import pandas as pd

data = {'y_Actual':    [1, 0, 0, 1, 0, 1, 0, 0, 1, 0, 1, 0],
        'y_Predicted': [1, 1, 0, 1, 0, 1, 1, 0, 1, 0, 0, 0]
        }
    
df = pd.DataFrame(data, columns = ['y_Actual', 'y_Predicted'])
print(df)
```

```python
confusion_matrix = pd.crosstab(df['y_Actual'], df['y_Predicted'], rownames=['Actual'], colnames=['Predicted'])
print (confusion_matrix)
```

The seaborn isused to plot the confusion matrix heatmap:

```python
import seaborn as sn
import matplotlib.pyplot as plt

sn.heatmap(confusion_matrix, annot=True, cmap="cubehelix")
```
